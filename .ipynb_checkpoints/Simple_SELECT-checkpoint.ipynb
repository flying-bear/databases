Далее создаем курсор для отправки запросов и получения результатов. `dictionary=True` показывает, что результаты надо возвращать в виде списка словарей, а не списка списков.

```python
cur = con.cursor(dictionary=True)
```

Выполним несложный запрос при помощи функции execute - выбрать все поля и все записи из таблицы articles. 

SELECT отвечает за выборку данных. * означает, что мы выбираем все поля, но здесь можно написать конкретные поля через запятую. Поля могут содержать имя таблицы, например, students.student_id. После FROM идут таблицы, из которых делается выборка.

```python
cur.execute("SELECT * FROM students")
```

Результаты запроса должны быть или извлечены с сервера (ниже будет fetch), или подтверждены (commit, например, для подтверждения вставки записей при помощи insert), или отменены (rollback, что-то пошло не так, отменяем последние операции до последнего подтверждения или отмены).

Совершенно не нужно вот так вот сразу отменять все результаты. На практике лучше пропустить эту строчку.

Тем более, что мы добавили ее потом просто чтобы посмотреть как работает.

```python
con.rollback()
```

Извлекаем все результаты запроса (напомню, это был SELECT) в переменную res. Она-то и будет списком словарей (потому что мы попросили об этом курсор).

```python
res = cur.fetchall()
```

Посмотрим на результаты выполнения запроса. В самом деле список словарей, поля записей названы также, как и в таблице.

Берем файл с именами, будем брать из него имена наших студентов. Возьмем первые 100 человек.

```python
with open ("/home/edward/projects/MNITI/NER Soft/OSM base/names/russian_names_government.txt", "rt") as names_file:
    names = names_file.readlines()[:100]
```

Немного облагородим эти имена.

```python
names2 = [name.strip().split() for name in names]
```

```python
names2
```

INSERT отвечает за добавление записей в таблицу.

`INSERT INTO таблица (поля через запятую) VALUES (значения через запятую)`

Можно добавить сразу много записей, пускаем их через запятые.

```python
%%time
req = """INSERT INTO students (surename,`name`,patronimyc, \
           `group`, birthday, speciality) VALUES """
reqs = []
for name in names2:
    reqs.append( """ ( '{}','{}','{}','{}','{}','{}')""".format(name[-1], name[0], "", 'GREAT-ONES', '2019-01-01', '00.00.00'))
req+=" , ".join(reqs)
cur.execute(req)
con.commit()
    
```

В конце обяательно говорим `COMMIT`, а то ничего не добавится.

```python
con.commit()
```
